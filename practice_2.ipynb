# Histograms and Bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(18,40,(100))
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("Ml Students age Histogram")
plt.xlabel("Students age catgory")
plt.ylabel("No. Students age")
plt.show()
```

```python
bins=[15,20,25,30,45]

plt.figure(figsize=(12,6))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype='bar',orientation='vertical',color=['m','y'],label=["Ml Student","py_student_age"])
#plt.hist(py_student_age,bins,rwidth=0.8,histtype='bar',orientation='vertical',color='y',label="PY Student")
plt.title("Ml Students age Histogram")
plt.xlabel("Students age catgory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[15,20,25,30,45]

plt.figure(figsize=(12,6))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype='bar',orientation='vertical',color=['m','y'],label=["Ml Student","py_student_age"])
#plt.hist(py_student_age,bins,rwidth=0.8,histtype='bar',orientation='vertical',color='y',label="PY Student")
plt.title("Ml Students age Histogram")
plt.xlabel("Students age catgory")
plt.ylabel("No. Students age")
plt.legend()
plt.grid(color='k', linestyle='-', linewidth=5)
plt.show()
```

```python
import matplotlib.pyplot as pt
import numpy as np
from matplotlib import style
```

```python
classes=["Python","R","AI","ML","DS"]
class1_students=[30,10,20,25,10]
class2_students=[40,5,20,20,10]
class3_students=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
'''
plt.bar(
    x,
    height,
    width=0.8,
    bottom=None,
    *,
    align='center',
    data=None,
    **kwargs,
)
'''
```

```python
plt.bar(classes,class1_students,width=0.2,align='edge',
        color='y',edgecolor='r',linewidth=5,alpha=0.8,linestyle="--",
        label="class 1 students",visible=True)#visible=False if you don't want see graph
```

```python
style.use("ggplot")
plt.figure( figsize=(12,6))
plt.bar(classes,class1_students,width=0.6,align='edge',
        color='k',edgecolor='r',linewidth=5,alpha=0.8,linestyle="--",
        label="class 1 students",visible=True)#visible=False if you don't want see graph
```

```python
style.use("ggplot")
plt.figure( figsize=(12,6))
plt.barh(classes,class1_students,align='edge',
        color='k',edgecolor='r',linewidth=5,alpha=0.8,linestyle="--",
        label="class 1 students",visible=True)#visible=False if you don't want see graph
```

```python
plt.figure( figsize=(12,6))
plt.bar(classes,class1_students,align='edge',
        color='k',edgecolor='r',linewidth=5,alpha=0.8,linestyle="--",
        label="class 1 students",visible=True)#visible=False if you don't want see graph
plt.title("Bar Chart of IAIP Class",fontsize=25)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```

```python
plt.figure( figsize=(12,6))

plt.bar(classes,class1_students,color='b',width=0.2,
        label="class 1 students")

plt.bar(classes,class2_students,color='g',width=0.2,
        label="class 1 students")

plt.title("Bar Chart of IAIP Class",fontsize=18)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```

```python
plt.figure( figsize=(12,6))

classes_index=np.arange(len(classes))

width=0.2

plt.bar(classes_index,class1_students,width,color='b',
        label="class 1 students")

plt.bar(classes_index+width,class2_students,width,color='g',
        label="class 1 students")

plt.bar(classes_index+width+width,class3_students,width,color='y',
        label="class 1 students")

plt.xticks(classes_index+width,classes,rotation=20)

plt.title("Bar Chart of IAIP Class",fontsize=18)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.show()
```

```python
plt.figure( figsize=(12,6))

classes_index=np.arange(len(classes))

width=0.2

plt.barh(classes_index,class1_students,width,color='b',
        label="class 1 students")

plt.barh(classes_index+width,class2_students,width,color='g',
        label="class 2 students")

plt.barh(classes_index+width+width,class3_students,width,color='y',
        label="class 3 students")

plt.yticks(classes_index+width,classes,rotation=20)

plt.title("Bar Chart of IAIP Class",fontsize=18)
plt.ylabel("Classes",fontsize=15)
plt.xlabel("No. of Students",fontsize=15)
plt.legend()
plt.show()
```
